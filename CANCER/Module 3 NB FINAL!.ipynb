## Disease Background:
- Cancer hallmark focus: Apoptosis
- Overview of hallmark: Apoptosis is the normal, planned death of damaged or old cells. It’s a natural process that keeps the body healthy by removing cells that could become harmful if they stay alive. When apoptosis happens correctly, it prevents damaged DNA from being copied and forming cancer. But when apoptosis is disrupted or fails to occur, those damaged cells survive and continue dividing, which can lead to tumor growth.

- Genes associated with hallmark to be studied (describe the role of each gene, signaling pathway, or gene set you are going to investigate):
TP53 (p53): A major tumor suppressor that detects genomic damage and can induce apoptosis. When mutated, which occurs in over half of human cancers, this pathway fails to activate cell death in response to stress.
BCL-2 family: A group of genes that regulate the intrinsic mitochondrial pathway of apoptosis. Anti-apoptotic members like BCL-2 and BCL-XL prevent the release of pro-death signals, while pro-apoptotic members like BAX and BAK promote cell death.
Caspases: The executioner proteases that carry out the apoptotic process by degrading key structural and regulatory proteins.
PI3K–AKT pathway: A signaling cascade that suppresses apoptosis and supports survival. Overactivation of this pathway in many tumors contributes to resistance to cell death.

This code will import all the libraries we need and set up basic plot styling.


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve
)

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6, 4)

```

This code will load the BRCA metadata, clean sample IDs, and keep only primary tumor samples.

```python
metadata_brca = pd.read_csv("GSE62944_metadata_BRCA.csv")

# Clean sample IDs (TCGA uses '-' not '.')
metadata_brca["sample_clean"] = metadata_brca["sample"].str.replace(".", "-", regex=False)

# Helper function to decode TCGA sample type from barcode
def sample_type_from_id(sample_id):
    if pd.isna(sample_id):
        return "Unknown"
    parts = str(sample_id).split('-')
    if len(parts) >= 4:
        code = parts[3][:2]
        if code in ["01", "03"]:
            return "Primary Tumor"
        elif code == "11":
            return "Solid Tissue Normal"
        else:
            return "Other"
    return "Unknown"

metadata_brca["sample_type"] = metadata_brca["sample_clean"].apply(sample_type_from_id)

# Keep only primary tumor samples
metadata_brca_tumor = metadata_brca[metadata_brca["sample_type"] == "Primary Tumor"].copy()
print("Number of BRCA tumor samples:", metadata_brca_tumor.shape[0])
```

This code will load the survival data, align patient barcodes, merge with tumor metadata,
and create a short vs long survival label based on a 4-year cutoff.

```python
# Load survival data and rename ID column
surv = pd.read_csv("subsampled_TCGA_CDR_survival.csv")
surv = surv.rename(columns={"Unnamed: 0": "bcr_patient_barcode"})

# Derive patient barcode (first 12 chars) from sample_clean
metadata_brca_tumor["bcr_patient_barcode"] = metadata_brca_tumor["sample_clean"].astype(str).str[:12]
surv["bcr_patient_barcode"] = surv["bcr_patient_barcode"].astype(str).str[:12]

# Merge metadata + survival
merged = metadata_brca_tumor.merge(surv, on="bcr_patient_barcode", how="inner")

# Drop samples with missing survival time
merged = merged.dropna(subset=["OS.time"])
print("Number of BRCA tumor samples with OS.time:", merged.shape[0])

# Define short survival (< 4 years) = 1, long survival (>= 4 years) = 0
threshold_days = 4 * 365  # 1460 days
merged["short_survival"] = (merged["OS.time"] < threshold_days).astype(int)

print("Short vs Long survival counts (4-year threshold):")
print(merged["short_survival"].value_counts())
```

This code will load the expression matrix, match samples with survival data, select apoptosis-related genes, and build the feature matrix X and label vector y.

```python
# Load expression data (log2 TPM)
expr = pd.read_csv("GSE62944_subsample_log2TPM.csv", index_col=0)
expr.columns = expr.columns.str.replace(".", "-", regex=False)

# Match expression columns to samples with survival data
final_samples = merged["sample_clean"].tolist()
expr_brca = expr[final_samples]

# Apoptosis-related genes of interest
genes_of_interest = ["TP53", "BCL2", "BAX", "CASP3"]

# Check if these gene names appear in the index
for g in genes_of_interest:
    if g not in expr_brca.index:
        hits = [x for x in expr_brca.index if g in x]
        print(f"{g} possible matches:", hits[:3])

# If needed, manually adjust to "GENE|ID" forms based on printed hits
# Example:
# genes_of_interest = ["TP53|7157", "BCL2|596", "BAX|581", "CASP3|836"]

# Build feature matrix (samples x genes)
X = expr_brca.loc[genes_of_interest].T
X = X.astype(float)

# Build label vector: short_survival (1) vs long_survival (0)
y = merged.set_index("sample_clean").loc[final_samples, "short_survival"]

print("Feature matrix shape:", X.shape)
print("Label distribution:\n", y.value_counts())
```

This code will run PCA on the standardized gene expression data and plot the first two principal components colored by survival group.

```python
# Standardize features for PCA
scaler_full = StandardScaler()
X_scaled_full = scaler_full.fit_transform(X)

# Fit PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_full)

# Build DataFrame for plotting
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=X.index)
pca_df['Survival_Group'] = y.map({0: 'Long', 1: 'Short'})

print("Explained variance by components:", pca.explained_variance_ratio_)

# Plot PCA colored by survival group
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=pca_df, x='PC1', y='PC2', hue='Survival_Group',
    palette={'Long': 'steelblue', 'Short': 'tomato'},
    s=80, alpha=0.8
)
plt.title('PCA of Apoptosis Gene Expression (BRCA Tumor Samples)')
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.legend(title='Survival Group')
plt.tight_layout()
plt.show()
```

This code will split the data into training and test sets and standardize the features for modeling.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("Train samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])

# Standardize on training data, apply same transform to test
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

This code will train a class-weighted logistic regression model on the scaled training data and compute predictions and probabilities.


```python
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(X_train_scaled, y_train)

# Predictions (class) and probabilities
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

y_train_proba = model.predict_proba(X_train_scaled)[:, 1]
y_test_proba = model.predict_proba(X_test_scaled)[:, 1]

print("Model training complete.")
```

This code will print accuracy and a full classification report (precision, recall, F1) for both the training and test sets.


```python
print("TRAIN SET SCORES")
print("Accuracy:", round(accuracy_score(y_train, y_train_pred), 3))
print(classification_report(y_train, y_train_pred))

print("\nTEST SET SCORES")
print("Accuracy:", round(accuracy_score(y_test, y_test_pred), 3))
print(classification_report(y_test, y_test_pred))
```

This code will create confusion matrix heatmaps for the training and test sets to show how often the model is right or wrong for each class.

```python
cm_train = confusion_matrix(y_train, y_train_pred)
cm_test = confusion_matrix(y_test, y_test_pred)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.heatmap(
    cm_train, annot=True, fmt="d", cmap="Blues", ax=axes[0],
    xticklabels=["Pred Long", "Pred Short"],
    yticklabels=["True Long", "True Short"]
)
axes[0].set_title("Train Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("True")

sns.heatmap(
    cm_test, annot=True, fmt="d", cmap="Blues", ax=axes[1],
    xticklabels=["Pred Long", "Pred Short"],
    yticklabels=["True Long", "True Short"]
)
axes[1].set_title("Test Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("True")

plt.tight_layout()
plt.show()
```

This code will compute AUC for train and test and plot the ROC curve for the test set.

```python
train_auc = roc_auc_score(y_train, y_train_proba)
test_auc = roc_auc_score(y_test, y_test_proba)

print("Train AUC:", round(train_auc, 3))
print("Test AUC:", round(test_auc, 3))

fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)

plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr, label=f"Model (AUC = {test_auc:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random (AUC = 0.5)")
plt.xlabel("False Positive Rate (1 - Specificity)")
plt.ylabel("True Positive Rate (Sensitivity)")
plt.title("ROC Curve – Short vs Long Survival (Test Set)")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

This code will test different decision thresholds (0.3, 0.5, 0.7) and show how accuracy, precision, recall, and F1 change on the test set.


```python
def evaluate_threshold(threshold):
    y_test_custom = (y_test_proba >= threshold).astype(int)
    print(f"\n=== Threshold = {threshold} ===")
    print("Accuracy:", round(accuracy_score(y_test, y_test_custom), 3))
    print(classification_report(y_test, y_test_custom))

for thr in [0.3, 0.5, 0.7]:
    evaluate_threshold(thr)
```

This code will compute McFadden's pseudo-R² to summarize overall model fit for the logistic regression.


```python
eps = 1e-15  # to avoid log(0)

# Log-likelihood of the fitted model
ll_full = np.sum(
    y_test * np.log(y_test_proba + eps) +
    (1 - y_test) * np.log(1 - y_test_proba + eps)
)

# Log-likelihood of a null model that always predicts mean(y)
p_null = np.mean(y_test)
ll_null = np.sum(
    y_test * np.log(p_null + eps) +
    (1 - y_test) * np.log(1 - p_null + eps)
)

r2_mcfadden = 1 - (ll_full / ll_null)
print(f"\nMcFadden's pseudo-R²: {r2_mcfadden:.4f}")
```

This code will print the logistic regression coefficients and plot them so you can see which genes push toward short vs long survival.


```python
coef_df = pd.DataFrame({
    "Gene": X.columns,
    "Coefficient": model.coef_[0]
}).sort_values("Coefficient", ascending=False)

print("\nLogistic Regression Coefficients:")
print(coef_df)

plt.figure(figsize=(6, 4))
sns.barplot(data=coef_df, x="Coefficient", y="Gene", palette="coolwarm")
plt.axvline(0, color="black", linewidth=1)
plt.title("Gene Expression Effect on Short Survival (4-Year Threshold)")
plt.xlabel("Logistic Regression Coefficient")
plt.ylabel("Gene")
plt.tight_layout()
plt.show()
```
