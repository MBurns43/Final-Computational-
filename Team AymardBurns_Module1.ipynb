* Societal determinants
There are many factors outside of biology and genetics that play a role in the risk of developing Alzheimer’s disease. These are social determinants of Alzheimer's, which include education, access to health care, the community, and their level of social connection. Education is one of the strongest determinants. People with lower levels of education are more likely to experience memory loss and cognitive decline later in life. For example, people 45+ without a high school diploma report more than twice the rate of worsening memory compared to people with a college degree. Education is believed to build what is called “cognitive reserve,” which helps protect the brain against changes that happen with age or disease. Health care access also makes a major difference. People who regularly see a doctor are more likely to catch and manage conditions like diabetes, hypertension, and heart disease, which are linked to a higher risk of dementia. Without consistent access to care, people may be diagnosed much later, which means symptoms are worse by the time they are treated. Additionally, the environment where someone lives also plays a role. Living in a neighborhood with limited access to safe sidewalks, healthy food, and reliable housing can increase risk for poor brain health. These conditions can limit physical activity, reduce good nutrition, and create stress, all of which contribute to cognitive decline. Social connections are another important factor. People who are isolated or lonely have a much higher chance of developing dementia. Staying socially engaged through friends, family, or community groups is protective for brain health, while isolation increases vulnerability.

* Diagnosis
The diagnosis of Alzheimer’s disease is done through a series of tests, as no one test alone can determine whether the patient has developed the condition. To begin, the patient would see a health care specialist, a neurologist (a doctor who is specialized on the brain and nervous system), or a geriatrician (a doctor specializing in the health care of older adults). At this stage, they would go over the symptoms the patient is experiencing, as well as medical and medicinal history. The doctor may also reach out to family or friends for answers about the patient and their behavior. Before testing, a physical exam is done to confirm that something else is not causing the symptoms. If the doctors cannot pin a different cause, then they begin the testing.
The next step is asking the patient a series of questions or having them complete certain actions to measure cognitive function such as memory, abstract thinking, problem solving, language, and more. After this, lab tests are performed, such as checking blood and urine samples. Neurological exams follow, where the patient is tested on reflexes, coordination, eye movement, speech, and sensation, and in some cases brain imaging is done. Brain imaging tests include an MRI (Magnetic Resonance Imaging), a CT (Computerized Tomography), and a PET (Positron Emission Tomography). There are also cerebrospinal fluid (CSF) tests, which look for changes in the CSF, the clear fluid that protects the brain and spinal cord, and measure biomarker levels such as tau and Abeta.
Other types of testing include the Ascertain Dementia 8 (AD8), the Saint Louis University Mental Status exam (SLUMS), the Informant Questionnaire on Cognitive Decline in the Elderly (IQCODE or s-IQCODE), the Functional Activities Questionnaire (FAQ), and the Mini-Mental State Exam (MMSE), which are cognitive, functional, and behavioral tests used to evaluate changes in behavior, memory, and thinking skills. There are also computerized cognitive tests, such as the Automated Neuropsychological Assessment Metrics (ANAM), the Cambridge Neuropsychological Test Automated Battery (CANTAB Mobile®), CognICA, Cognigram, Cognivue, and Cognision. The last type of testing commonly used is genetic testing, since there are genetic determinants of Alzheimer’s. There is the blood test for the strongest risk gene, APOE-e4, but there is also testing for genes that cause ADAD, autosomal dominant Alzheimer’s disease.

* Disease progression & prognosis
Alzheimer’s is a progressive disease, meaning it develops slowly and worsens over time. According to the Mayo Clinic, there are five stages associated with its progression: preclinical Alzheimer’s disease, mild cognitive impairment due to Alzheimer’s disease (MCI), mild dementia due to Alzheimer’s disease, moderate dementia due to Alzheimer’s disease, and severe dementia due to Alzheimer’s disease.
The preclinical stage is when changes in the brain begin to occur before any symptoms are noticeable. This stage can last for years or even decades, and the only way to detect it is through testing such as brain imaging, which may show signs of biomarkers. Genetic testing and blood samples can also be helpful during this stage. In the mild cognitive impairment stage, the patient experiences slight changes in memory and thinking skills, such as forgetting conversations or events. At this point, awareness of time often declines, decision-making becomes more difficult, and judgment may be impaired, though not all patients in this stage necessarily have Alzheimer’s.
Mild dementia is typically the stage when Alzheimer’s is diagnosed, as memory and thinking troubles become very apparent and start affecting daily life. Symptoms include memory loss of recent events, difficulty problem solving, personality changes, poor judgment, communication issues, and frequently misplacing items. As the disease progresses to moderate dementia, these existing symptoms worsen and new ones appear. Patients may require help completing daily tasks, forget the names of close family members or even their own, struggle to learn new things, experience mood swings or agitation, and develop trouble sleeping.
In the final stage, severe dementia, patients lose many motor functions, including the ability to walk, sit, or eat without assistance, and require full-time care. They may lose bladder and bowel control, be unable to hold conversations, and only manage a few words. At this stage, they are often unaware of their surroundings and become more susceptible to infections.
Alzheimer’s has no known cure, and once diagnosed, patients typically live three to eleven years, though some live more than twenty depending on the stage at diagnosis. Common causes of death in people with Alzheimer’s include pneumonia, dehydration, malnutrition, falls, and infections.

* Continuum of care providers
There are different kinds of providers and services involved in Alzheimer’s care over time. These providers help from early stages through late stages of dementia. The continuum includes things like risk assessment, early diagnosis, ongoing care, caregiving support, and community‐based services. Community-based organizations often are the first place people turn when they notice memory or thinking problems. They can help by giving information, connecting families to care, and navigating local resources. Then medical providers like primary care doctors, neurologists, geriatricians, neuropsychologists, and specialists in mental health become involved. They handle diagnostic evaluations, treatment planning, and follow-up support. After diagnosis, ongoing support is crucial. Health care providers offer post-diagnostic care and help with managing symptoms, both behavioral and cognitive, planning for future care needs, and coordinating with caregivers and family. However, caregivers themselves also need support, emotional, practical, and educational support. Federal resources also help providers do their work. There are training programs, guidelines, and toolkits to help doctors, nurses, therapists, and other professionals understand dementia, diagnose it early, and provide quality care. There are also policies and systems-level efforts to improve how care is delivered across settings (clinics, homes, long-term care) so that people with Alzheimer’s and their families get more consistent, coordinated care.

## Data-Set: 
This project uses two datasets: Updated Meta Data (Set 1) and Updated Luminex (Set 2). Set 1 provides demographic, clinical, genetic, cognitive, and neuropathological information for 84 human brain donors. Demographics include sex, race, age at death (years), and years of education. Genetic information includes the APOE genotype (e.g., 3/3, 3/4, 4/4). Clinical and cognitive data describe dementia status, age of symptom onset and diagnosis, history of head injury, neuroimaging records, and other diseases. Cognitive performance was assessed with standardized tests such as the Cognitive Abilities Screening Instrument (CASI, 0–100), the Mini-Mental State Examination (MMSE, 0–30), and the Montreal Cognitive Assessment (MOCA, 0–30), along with the number of months between the last test and death. Neuropathological measures include Braak stage** (I–VI, spread of tau tangles), CERAD score (density of amyloid plaques), Thal phase (1–5, spread of plaques), and CAA rating (amyloid in blood vessel walls). Other measures capture whether the donor had LATE (TDP-43 proteinopathy, staged 1–4), Lewy body disease, microinfarcts (tiny strokes), and vascular conditions such as atherosclerosis and arteriolosclerosis. Tissue quality variables are also included, such as postmortem interval (hours), fresh brain weight (grams), and brain pH.
Set 2 contains quantitative biomarker data from the same 84 donors, measured using Luminex® multiplex immunoassay technology, which allows precise measurement of multiple proteins within a single sample. The biomarkers measured are amyloid-beta 40 (ABeta40), amyloid-beta 42 (ABeta42), total tau (tTau), and phosphorylated tau (pTau), which are hallmark proteins associated with Alzheimer’s pathology. Concentrations are reported in picograms per microgram (pg/µg) of protein, providing standardized quantitative units.
Together, these datasets allow for the integration of demographic, cognitive, and neuropathological data (Set 1) with molecular biomarker levels (Set 2), enabling comprehensive analyses of Alzheimer’s disease progression. This is the Patient class definition with the class methods. There is a separate .py file called patient.py containing the class definition

```python
#1) BUILD OUR CLASS BY COMBINING TWO .csv FILES OF DATA
 
import csv
import warnings
import matplotlib.pyplot as plt

class Patient: 

    all_patients = []

    death_age = []

    education_lvl = {}

    athero_lvl = {}

    arterio_lvl = {}

    def __init__(self, DonorID, ABeta40: float , ABeta42: float, tTau: float, pTau: float):

        self.DonorID = DonorID
        self.ABeta40 = ABeta40
        self.ABeta42 = ABeta42
        self.tTau = tTau
        self.pTau = pTau
        self.sex = None
        self.death_age = None
        self.ed_lvl = None
        self.cog_stat = None
        self.age_symp_on = None
        self.age_diag = None 
        self.head_inj = None
        self.thal_score = None
        self.athero = None
        self.arterio = None
        Patient.all_patients.append(self)

    def __repr__(self):
        return f"{self.DonorID} | sex: {self.sex} | ABeta40 {self.ABeta40} | tTau {self.tTau} | pTau {self.pTau} | Death Age {self.death_age} | Thal Score {self.thal_score} | Atherosclerosis {self.athero} | Arteriolosclerosis {self.arterio}"

    def get_id(self):
        return self.DonorID

    def get_ABeta42(self):
        return self.ABeta42
    
    def get_ABeta40(self):
        return self.ABeta40
    
    def get_thal(self):
        return self.thal_score
    
    def get_death_age(self):
        return self.death_age
    
    def get_athero(self):
        return self.athero
    
    def get_arterio(self):
        return self.arterio


    @classmethod
    def combine_data(cls, filename: str):
        with open(filename, encoding="utf8") as f:
            reader = csv.DictReader(f)
            rows_of_patients = list(reader)
            # Build a lookup by Donor ID
            meta_by_id = {row["Donor ID"]: row for row in rows_of_patients}
            for patient in cls.all_patients:
                row = meta_by_id.get(patient.DonorID)
                if row is None:
                    warnings.warn(f"Donor ID {patient.DonorID} not found in metadata CSV.")
                    continue
                if row["Sex"] != "":
                    patient.sex = row["Sex"]
                if row["Age at Death"] != "":
                    patient.death_age = int(row["Age at Death"])
                if row["Highest level of education"] != "":
                    patient.ed_lvl = row["Highest level of education"]
                if row["Cognitive Status"] != "":
                    patient.cog_stat = row["Cognitive Status"]
                if row["Age of onset cognitive symptoms"] != "":
                    patient.age_symp_on = int(row["Age of onset cognitive symptoms"])
                if row["Age of Dementia diagnosis"] != "":
                    patient.age_diag = int(row["Age of Dementia diagnosis"])
                if row["Known head injury"] != "":
                    patient.head_inj = row["Known head injury"]
                if row["Atherosclerosis"] != "":
                    patient.athero = row["Atherosclerosis"]
                if row["Arteriolosclerosis"] != "":
                    patient.arterio = row["Arteriolosclerosis"]
                if row["Thal"] != "":
                    patient.thal_score = int(row["Thal"].split()[1])
    @classmethod
    def instantiate_from_csv(cls, filename: str, other_file: str):
    #open csv and create list of all rows
        with open(filename, encoding="utf8") as f:
            reader = csv.DictReader(f)
            rows_of_patients = list(reader)
            #for line in csv create object
            for row in rows_of_patients:
                Patient(
                    DonorID = row['Donor ID'],
                    ABeta40 = float(row['ABeta40 pg/ug']),
                    ABeta42 = float(row['ABeta42 pg/ug']),
                    tTau = float(row['tTAU pg/ug']),
                    pTau = float(row['pTAU pg/ug'])
                )
            Patient.all_patients.sort(key = Patient.get_id)
            Patient.combine_data(other_file)
    #4) BUILD DICTIONARIES TO DO SORTING AND SUB-SORTING

    @classmethod
    def sort_ed(cls):
        for patient in cls.all_patients:
            cls.education_lvl.update({patient.ed_lvl: []})
        for patient in cls.all_patients:
            cls.education_lvl[patient.ed_lvl].append(patient)

    @classmethod
    def subsort_thal(cls):
        for key in cls.education_lvl:
            values = cls.education_lvl.get(key)
            values.sort(key = cls.get_thal)
            cls.education_lvl.update({key: values})

    @classmethod
    def sort_athero(cls):
        for patient in cls.all_patients:
            cls.athero_lvl.update({patient.athero: []})
        for patient in cls.all_patients:
            cls.athero_lvl[patient.athero].append(patient)
    
    @classmethod
    def sort_arterio(cls):
        for patient in cls.all_patients:
            cls.arterio_lvl.update({patient.arterio: []})
        for patient in cls.all_patients:
            cls.arterio_lvl[patient.arterio].append(patient)

    @classmethod
    def subsort_ABeta40(cls):
        for key in cls.athero_lvl:
            values = cls.athero_lvl.get(key)
            values.sort(key=lambda patient: patient.get_ABeta40())
            cls.athero_lvl.update({key: values})

    @classmethod
    def filter(cls, list, ABeta40:float ="any", ABeta42:float ="any", tTau:float= "any", pTau:float ="any", sex:str ="any", death_age:int ="any", ed_lvl:str ="any", cog_stat:str ="any", age_symp_on:int ="any", age_diag:int ="any", head_inj:str ="any", athero:str ="any", arterio:str ="any", thal_score:int ="any"):
        all_patients = list
        remove_list = []
        attr_list = (
                    ABeta40,
                    ABeta42,
                    tTau,
                    pTau,
                    sex,
                    death_age,
                    ed_lvl,
                    cog_stat,
                    age_symp_on,
                    age_diag,
                    head_inj,
                    athero,
                    arterio,
                    thal_score
                    )
        attr_name = (
                    "ABeta40",
                    "ABeta42",
                    "tTau",
                    "pTau",
                    "sex",
                    "death_age",
                    "ed_lvl",
                    "cog_stat",
                    "age_symp_on",
                    "age_diag",
                    "head_inj",
                    "athero",
                    "arterio",
                    "thal_score"
                    )
        for attr in range(len(attr_list)):
            if attr_list[attr] != "any":
                for patient in all_patients:
                    if getattr(patient, attr_name[attr], None) != attr_list[attr]:
                        remove_list.append(patient)
                all_patients = [patient for patient in all_patients if patient not in remove_list]
                remove_list.clear()
        
        return all_patients



    
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# ---- Gather ABeta40 by atherosclerosis level ----
levels = ["None", "Mild", "Moderate", "Severe"]

group_vals = {lvl: [p.ABeta40 for p in Patient.filter(Patient.all_patients, athero=lvl)]
              for lvl in levels}

# Optionally, drop empty groups to avoid ANOVA/Tukey errors
nonempty_levels = [lvl for lvl in levels if len(group_vals[lvl]) > 0]
groups_for_anova = [group_vals[lvl] for lvl in nonempty_levels]

# ---- Bar chart with means & SDs (optional) ----
means = [np.mean(group_vals[lvl]) if len(group_vals[lvl]) > 0 else 0 for lvl in levels]
stdevs = [np.std(group_vals[lvl]) if len(group_vals[lvl]) > 0 else 0 for lvl in levels]

plt.figure(figsize=(7,5))
plt.bar(levels, means, yerr=stdevs, capsize=8)
plt.ylabel("ABeta40")
plt.title("ABeta40 by Atherosclerosis Level")
plt.show()

# ---- One-way ANOVA (SciPy quick test) ----
if len(groups_for_anova) >= 2:
    f_stat, p_val = stats.f_oneway(*groups_for_anova)
    print(f"ANOVA F-statistic: {f_stat:.4f}")
    print(f"ANOVA p-value:     {p_val:.4g}")
else:
    print("Not enough non-empty groups for ANOVA.")

# ---- ANOVA via statsmodels (gives full table) ----
# Build long-form DataFrame for statsmodels & Tukey
vals = []
labs = []
for lvl in nonempty_levels:
    vals.extend(group_vals[lvl])
    labs.extend([lvl] * len(group_vals[lvl]))

df = pd.DataFrame({"ABeta40": vals, "athero": labs})

if df["athero"].nunique() >= 2:
    model = ols("ABeta40 ~ C(athero)", data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("\nANOVA table (statsmodels):")
    print(anova_table)
else:
    print("Not enough levels in data for statsmodels ANOVA.")

# ---- Tukey HSD post-hoc (which pairs differ) ----
# Requires at least 2 groups and enough observations
if df["athero"].nunique() >= 2:
    tukey = pairwise_tukeyhsd(endog=df["ABeta40"], groups=df["athero"], alpha=0.05)
    print("\nTukey HSD results:")
    print(tukey.summary())
else:
    print("Not enough levels for Tukey HSD.")

```
