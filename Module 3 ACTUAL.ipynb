## Disease Background:
- Cancer hallmark focus: Apoptosis
- Overview of hallmark: Apoptosis is the normal, planned death of damaged or old cells. It’s a natural process that keeps the body healthy by removing cells that could become harmful if they stay alive. When apoptosis happens correctly, it prevents damaged DNA from being copied and forming cancer. But when apoptosis is disrupted or fails to occur, those damaged cells survive and continue dividing, which can lead to tumor growth.

- Genes associated with hallmark to be studied (describe the role of each gene, signaling pathway, or gene set you are going to investigate):
TP53 (p53): A major tumor suppressor that detects genomic damage and can induce apoptosis. When mutated, which occurs in over half of human cancers, this pathway fails to activate cell death in response to stress.
BCL-2 family: A group of genes that regulate the intrinsic mitochondrial pathway of apoptosis. Anti-apoptotic members like BCL-2 and BCL-XL prevent the release of pro-death signals, while pro-apoptotic members like BAX and BAK promote cell death.
Caspases: The executioner proteases that carry out the apoptotic process by degrading key structural and regulatory proteins.
PI3K–AKT pathway: A signaling cascade that suppresses apoptosis and supports survival. Overactivation of this pathway in many tumors contributes to resistance to cell death.

## Methods:

We used logistic regression, a supervised learning method that predicts whether a sample is tumor or normal based on gene expression. The model finds patterns in the data by adjusting how much weight it gives each gene. It measures how well it’s doing using log loss, and stops training once it can’t improve accuracy any further.

## Data-Set: 
The data from this project comes from two papers building upon The Cancer Gene Atlas [TCGA], which synthesizes data from various tumors, providing a comprehensive database of information about the cancer and the patient. These papers are the Pan-Cancer Clinical Data Resource (Liu et al., 2018) and the Reprocessed TCGA RNA-Sequencing Dataset (Rahman et al., 2015). The first paper, Pan-Cancer Clinical Data Resource, is a paper breaking down around 11,000 cases of 33 different types of cancers, mentioning cancer progression, patient demographics, and survival. The Reprocessed TCGA RNA-Sequencing Dataset used the Rsubread Pipeline to process gene-expression data from 9,264 tumor and 741 normal samples. Of these 10,005 data points, 23 cancer types are represented. The data created from the Rsubread Pipeline created integer values in units of FPKM and TPM.

```python
# %% ==========================================
# SECTION 1 — DATA
# ==========================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

# --- Load metadata and filter for BRCA ---
metadata = pd.read_csv("GSE62944_metadata.csv")
brca_metadata = metadata.loc[metadata['cancer_type'] == 'BRCA']
brca_metadata.to_csv("GSE62944_metadata_BRCA.csv", index=False)

# --- Clean sample IDs ---
metadata_brca = pd.read_csv("GSE62944_metadata_BRCA.csv")
metadata_brca["sample_clean"] = metadata_brca["sample"].str.replace(".", "-", regex=False)

# --- Identify tumor samples ---
def sample_type_from_id(sample_id):
    if pd.isna(sample_id):
        return "Unknown"
    parts = str(sample_id).split('-')
    if len(parts) >= 4:
        code = parts[3][:2]
        if code in ["01", "03"]:
            return "Primary Tumor"
        elif code == "11":
            return "Solid Tissue Normal"
        else:
            return "Other"
    return "Unknown"

metadata_brca["sample_type"] = metadata_brca["sample_clean"].apply(sample_type_from_id)
metadata_brca_tumor = metadata_brca[metadata_brca["sample_type"] == "Primary Tumor"].copy()
print("Number of BRCA tumor samples:", metadata_brca_tumor.shape[0])

# --- Load survival data ---
surv = pd.read_csv("subsampled_TCGA_CDR_survival.csv")
surv = surv.rename(columns={"Unnamed: 0": "bcr_patient_barcode"})
metadata_brca_tumor["bcr_patient_barcode"] = metadata_brca_tumor["bcr_patient_barcode"].astype(str).str[:12]
surv["bcr_patient_barcode"] = surv["bcr_patient_barcode"].astype(str).str[:12]

# --- Merge metadata and survival data ---
merged = metadata_brca_tumor.merge(surv, on="bcr_patient_barcode", how="inner")
merged = merged.dropna(subset=["OS.time"])
print("Number of BRCA tumor samples with OS.time:", merged.shape[0])

# --- Label survival outcome ---
threshold_days = 4 * 365
merged["short_survival"] = (merged["OS.time"] < threshold_days).astype(int)
print("Short vs Long survival counts:")
print(merged["short_survival"].value_counts())

# --- Load gene expression data ---
expr = pd.read_csv("GSE62944_subsample_log2TPM.csv", index_col=0)
expr.columns = expr.columns.str.replace(".", "-", regex=False)

# --- Match samples with survival data ---
final_samples = merged["sample_clean"].tolist()
expr_brca = expr[final_samples]

# --- Select apoptosis-related genes ---
genes_of_interest = ["TP53", "BCL2", "BAX", "CASP3"]

# Check gene naming consistency
for g in genes_of_interest:
    if g not in expr_brca.index:
        hits = [x for x in expr_brca.index if g in x]
        print(f"{g} possible matches:", hits[:3])

# --- Build feature matrix and target ---
X = expr_brca.loc[genes_of_interest].T
X = X.astype(float)
y = merged.set_index("sample_clean").loc[final_samples, "short_survival"]

print("Feature matrix shape:", X.shape)
print("Label distribution:\n", y.value_counts())

# --- Split data ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print("Train samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])

# --- Standardize and run PCA ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# PCA DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=X.index)
pca_df['Survival_Group'] = y.map({0: 'Long', 1: 'Short'})

# PCA Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df, x='PC1', y='PC2', hue='Survival_Group',
    palette={'Long': 'steelblue', 'Short': 'tomato'}, s=80, alpha=0.8
)
plt.title('PCA of Apoptosis Gene Expression (BRCA Tumor Samples)')
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.legend(title='Survival Group')
plt.tight_layout()
plt.show()
print("Explained variance by components:", pca.explained_variance_ratio_)
```

```python

# %% ==========================================
# SECTION 2 — VERIFY / VALIDATE
# ==========================================

# --- Logistic Regression ---
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# --- Model Performance ---
acc = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", round(acc, 3))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix Visualization
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Long Survival", "Short Survival"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix — Logistic Regression Model")
plt.show()

# --- Feature Importance ---
coef_df = pd.DataFrame({
    "Gene": X.columns,
    "Coefficient": model.coef_[0]
}).sort_values("Coefficient", ascending=False)

print("\nLogistic Regression Coefficients:")
print(coef_df)

# Barplot for top coefficients
plt.figure(figsize=(10, 6))
sns.barplot(data=coef_df, x="Coefficient", y="Gene", palette="coolwarm")
plt.axvline(0, color="black", linewidth=1)
plt.title("Gene Expression Influence on Survival Outcome")
plt.xlabel("Logistic Regression Coefficient")
plt.ylabel("Gene")
plt.tight_layout()
plt.show()

# --- PCA Plot by Predicted Group ---
pca_df['Predicted_Group'] = model.predict(X_scaled)
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df, x='PC1', y='PC2',
    hue='Predicted_Group', palette={0: 'steelblue', 1: 'tomato'},
    s=80, alpha=0.8
)
plt.title('PCA Plot Colored by Predicted Survival Group')
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.legend(title='Predicted Group')
plt.tight_layout()
plt.show()


```
